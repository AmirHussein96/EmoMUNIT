```python
import math
import numpy as np
import pandas as pd
from collections import defaultdict, Counter
```

```python
df = pd.read_csv("c/Batch_3588968_batch_results.csv", header=None)
print(df)
```

```python
d_emo = {'Angry':0, 'Happy':1, 'Neutral':2, 'Sad':3}
d_qua = {'Bad':0, 'Noisy':1, 'Fair':2, 'Good':3, 'Excellent':4}
d_sim = {'Different speakers':0, 'Hard to say':1, 'Similar speaker':2, 'Same speaker':3}
```

```python
emotion = defaultdict(list)
quality = defaultdict(list)
similarity = defaultdict(list)
for i in range(len(df)):
    idx = df[0][i].split('.')[0]
    if idx.isdigit():
        emo = df[1][i] if type(df[1][i])==str else df[2][i]
        qua = df[3][i] if type(df[3][i])==str else df[4][i]
        sim = df[5][i] if type(df[5][i])==str else df[6][i]
        try:
            emotion[idx].append(d_emo[emo])
            quality[idx].append(d_qua[qua])
            similarity[idx].append(d_sim[sim])
        except:
            print([df[x][i] for x in range(7)])
```

```python
r = ['ang2hap','ang2neu','ang2sad','hap2ang','hap2neu','hap2sad','neu2ang','neu2hap','neu2sad','sad2ang','sad2hap','sad2neu']
res_emo = {}
for c in r: res_emo[c] = [0]*4
```

```python
data, res = emotion, res_emo
for k in data:
    if 0 <= int(k) <= 18:
        for i in range(4): res['ang2hap'][data[k][i]] += 1
    if 19 <= int(k) <= 37:
        for i in range(4): res['ang2neu'][data[k][i]] += 1
    if 38 <= int(k) <= 56:
        for i in range(4): res['ang2sad'][data[k][i]] += 1
    if 57 <= int(k) <= 88:
        for i in range(4): res['hap2ang'][data[k][i]] += 1
    if 89 <= int(k) <= 120:
        for i in range(4): res['hap2neu'][data[k][i]] += 1
    if 121 <= int(k) <= 152:
        for i in range(4): res['hap2sad'][data[k][i]] += 1
    if 153 <= int(k) <= 195:
        for i in range(4): res['neu2ang'][data[k][i]] += 1
    if 196 <= int(k) <= 238:
        for i in range(4): res['neu2hap'][data[k][i]] += 1
    if 239 <= int(k) <= 281:
        for i in range(4): res['neu2sad'][data[k][i]] += 1
    if 282 <= int(k) <= 295:
        for i in range(4): res['sad2ang'][data[k][i]] += 1
    if 296 <= int(k) <= 309:
        for i in range(4): res['sad2hap'][data[k][i]] += 1
    if 310 <= int(k) <= 323:
        for i in range(4): res['sad2neu'][data[k][i]] += 1
```

```python
# count to percentage
for k in res:
    total = float(sum(res[k]))
    for i in range(4):
        res[k][i] = int(res[k][i] / total * 1000)/10
```
